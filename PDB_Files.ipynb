```python
import os
import pandas as pd
import requests

# Función para obtener y guardar la estructura PDB desde AlphaFold
def get_alphafold_structure(uniprot_id, binding_protein_name, save_path):
    url = f"https://alphafold.ebi.ac.uk/api/prediction/{uniprot_id}"
    response = requests.get(url)

    if response.status_code == 200:
        result = response.json()
        pdb_url = result[0]['pdbUrl']
        pdb_response = requests.get(pdb_url)

        if pdb_response.status_code == 200:
            # Crear el nombre del archivo PDB dentro de la carpeta especificada
            pdb_filename = f"{binding_protein_name}.pdb"
            save_file_path = os.path.join(save_path, pdb_filename)

            # Guardar el archivo PDB en la carpeta ProteinasPDB
            with open(save_file_path, "w") as pdb_file:
                pdb_file.write(pdb_response.text)
            print(f"Estructura PDB guardada como {save_file_path}")
        else:
            print(f"Error al descargar el archivo PDB para {binding_protein_name}")
    else:
        print(f"Error al obtener la predicción de AlphaFold para {uniprot_id}")

# Leer el archivo Excel para obtener el ID de UniProt y el nombre de la proteína de unión
def read_data_from_excel(file_path, uniprot_column, binding_protein_column):
    df = file_path
    if uniprot_column in df.columns and binding_protein_column in df.columns:
        return df[[uniprot_column, binding_protein_column]]
    else:
        raise ValueError(f"Las columnas '{uniprot_column}' o '{binding_protein_column}' no se encontraron en el archivo Excel")

# Ruta de guardado para los archivos PDB
save_path = "/home/nfernandez/unit/ProteinasPDB"

# Crear la carpeta si no existe
os.makedirs(save_path, exist_ok=True)

# Cargar el archivo Excel (asegúrate de que 'protein_sequences' sea una ruta válida de tu archivo)
file_path = protein_sequences
uniprot_column = "UniProtID"  # Asegúrate de que este sea el nombre correcto de la columna en el archivo
binding_protein_column = "Protein"  # Nombre de la proteína de unión

# Leer los datos del archivo Excel
data = read_data_from_excel(file_path, uniprot_column, binding_protein_column)

# Obtener la estructura para cada par de ID de UniProt y nombre de proteína de unión
for index, row in data.iterrows():
    uniprot_id = row[uniprot_column]
    binding_protein_name = row[binding_protein_column]

    if uniprot_id:
        get_alphafold_structure(uniprot_id, binding_protein_name, save_path)

```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
import os

df5 = result_df
df = df5
pdb_folder = '/home/nfernandez/unit/LigandosPDB'

# Crear la carpeta de salida si no existe
if not os.path.exists(pdb_folder):
    os.makedirs(pdb_folder)

# Función para guardar PDB a partir de SMILES
def save_pdb(smiles, mol_id, folder):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        # Verificar cuántos fragmentos tiene la molécula (sales u otros)
        frags = Chem.GetMolFrags(mol, asMols=True, sanitizeFrags=True)

        # Si hay más de un fragmento, nos quedamos con el más grande
        if len(frags) > 1:
            mol = max(frags, key=lambda frag: frag.GetNumAtoms())

        try:
            # Kekuliza, verifica las valencias, establece la aromaticidad, conjugación e hibridación
            Chem.SanitizeMol(mol)

            # Añadir hidrógenos
            mol = Chem.AddHs(mol)

            # Embebe la molécula en 3D
            AllChem.EmbedMolecule(mol)
            AllChem.UFFOptimizeMolecule(mol)

            # Guardar la molécula como archivo PDB
            pdb_filename = f'{folder}/{mol_id}.pdb'
            with open(pdb_filename, 'w') as f:
                f.write(Chem.MolToPDBBlock(mol))
            print(f"Archivo PDB guardado como {pdb_filename}")
        except Exception as e:
            print(f"Error al procesar el SMILES: {smiles} - {e}")
    else:
        print(f"Error al convertir el SMILES: {smiles}")

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    save_pdb(row['Smiles'], row['id'], pdb_folder)
```

```python
import os
from openbabel import pybel

# Ruta de la carpeta con los archivos PDB
input_folder = '/home/nfernandez/unit/LigandosPDB'
output_folder = '/home/nfernandez/unit/LigandosMOL'

# Asegúrate de que la carpeta de salida exista
os.makedirs(output_folder, exist_ok=True)

# Procesar cada archivo PDB en la carpeta
for filename in os.listdir(input_folder):
    if filename.endswith('.pdb'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename.replace('.pdb', '.mol2'))

        # Leer el archivo PDB
        mol = [m for m in pybel.readfile('pdb', input_path)][0]
        mol.addh()  # Agregar hidrógenos

        # Guardar el archivo como MOL2
        out = pybel.Outputfile('mol2', output_path, overwrite=True)
        out.write(mol)
        out.close()

        print(f"Archivo convertido: {filename} → {output_path}")
```
