```python
VOC_Ki = dfs_info_transformed.copy()

data = []

# Iterar sobre las filas de VOC_Ki
for index, row in VOC_Ki.iterrows():
    compound_name = row['Compound name']
    #smiles = row['Smiles']
    for column in VOC_Ki.columns[1:]:  # Iterar sobre las columnas después de 'Compound name'
        affinity = row[column]
        if affinity != 10000:
            data.append([compound_name, column, affinity])

# Crear un DataFrame a partir de los datos
new_df = pd.DataFrame(data, columns=['Compound name', 'Protein', 'Binding_Affinity'])

# Mostrar el nuevo DataFrame
new_df
```

```python
# Merge the two DataFrames based on the 'id' column
merged_df123 = pd.merge(new_df, dfs_info_transformed_grouped1[['Compound name', 'Smiles']], on='Compound name', how='left')

# Concatenate the 'Smiles' values from dfs_info_transformed_grouped1 into the original df
new_df['Smiles'] = merged_df123['Smiles']

# Display the updated DataFrame
new_df
```

```python
df = new_df
# Diccionario para almacenar los IDs asignados a cada SMILES
id_dict = {}
id_counter = 1  # Contador inicial para los IDs

# Asignar un ID numérico a cada SMILES, reutilizando el mismo ID si ya fue asignado antes
def assign_id(smiles):
    global id_counter
    if smiles not in id_dict:
        id_dict[smiles] = id_counter  # Asignar un nuevo ID si el SMILES no tiene uno
        id_counter += 1  # Incrementar el contador para el siguiente ID
    return id_dict[smiles]

# Crear una nueva columna 'id' con IDs asignados
df['id'] = df['Smiles'].apply(assign_id)

df
```

```python
merged_df = pd.merge(df, df_filtrado_pbps_gobps, left_on='Protein', right_on='Binding Protein Name', how='left')

# Select the desired columns from the merged DataFrame
result_df = merged_df[['id', 'Smiles', 'Protein', 'UniProtID', 'AA Sequence W/O signal peptide', 'Alphafold',  'Binding_Affinity']]

# Display the result
result_df
```

```python
#concatenar datos de diferentes dataframe
def concatenate_compound_names(smiles):
  """
  Concatena los nombres de los compuestos para un SMILES dado.

  Args:
    smiles: El SMILES para el que se desea concatenar los nombres de los compuestos.

  Returns:
    Una cadena con los nombres de los compuestos concatenados, separados por comas.
  """
  compound_names = dfs_info_transformed_grouped1[dfs_info_transformed_grouped1['Smiles'] == smiles]['Compound name'].tolist()
  return ', '.join(compound_names)

# Aplica la función a la columna Smiles y crea una nueva columna 'Compound Names'
result_df['Compound Name'] = result_df['Smiles'].apply(concatenate_compound_names)
cols = result_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
result_df = result_df[cols]
result_df
```
